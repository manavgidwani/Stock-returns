```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['BHARTIARTL.NS','HDFC.NS','AXISBANK.NS','RELIANCE.NS','ADANIENT.NS','TCS.NS','^NSEI']
BHARAT = pd.DataFrame()
for b in tickers:
    BHARAT[b] = wb.DataReader(b, data_source= 'yahoo', start = '2009-1-1')['Adj Close']
    
BHARAT.tail()
```

```python
BHARAT.iloc[0]
```

```python
BHARAT/BHARAT.iloc[0]*100
```

```python
(BHARAT/BHARAT.iloc[0]*100).plot(figsize=(15,10))
```

```python
log_return = np.log(1+BHARAT.pct_change(1))
log_return
```

```python
log_return.mean()*250
```

```python
log_return.std()*250**0.5
```
